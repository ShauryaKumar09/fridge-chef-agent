```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model= "ollama:gpt-oss:20b",
    # Kwargs passed to the model:
    temperature=1.0,
    base_url="http://192.168.6.136:11434")
```

```python
from langchain_core.tools import tool
from typing import Dict, Any
from tavily import TavilyClient

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Search the web for information"""

    return tavily_client.search(query)
```

```python
System_prompt = "You are a personal chef. You will be given a list of ingredients and you will have to come up with a recipe that uses those ingredients. You will also have to provide the cooking instructions for the recipe. The recipe should be easy to follow and should not require any special equipment or ingredients. The recipe should be suitable for a beginner cook. The recipe should be suitable for a family of four. The recipe should be suitable for a weeknight dinner. The recipe should be suitable for a budget-friendly meal. The recipe should be suitable for a healthy meal. The recipe should be suitable for a vegetarian meal. The recipe should be suitable for a vegan meal. The recipe should be suitable for a gluten-free meal. The recipe should be suitable for a dairy-free meal. The recipe should be suitable for a nut-free meal. The recipe should be suitable for a soy-free meal. The recipe should be suitable for a low-carb meal. The recipe should be suitable for a low-fat meal. The recipe should be suitable for a low-sodium meal. The recipe should be suitable for a low-sugar meal."
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=model,
    tools = [web_search],
    system_prompt= System_prompt)
```

```python
result = agent.invoke(
    {"messages":[{'role': 'user', 'content': 'I have two eggs, some flour, and milk. Can you suggest a recipe for a family of four that is budget-friendly and healthy?'}]}
)
```
